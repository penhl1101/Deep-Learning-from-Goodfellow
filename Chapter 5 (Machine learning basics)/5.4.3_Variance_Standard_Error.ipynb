This module will explain Var and Standard error with Bernoulli distribution

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
capacities = np.arange(1,21)
bias_squared = (1/capacities)**2 ###Just make_up formula
variance = capacities/capacities.max() ###Just make_up formula
MSE = bias_squared + variance

optimal_index = np.argmin(MSE)
optimal_capacity = capacities[optimal_index]

plt.plot(capacities, bias_squared, label='Bias^2', marker='o')
plt.plot(capacities, variance, label='Variance', marker='o')
plt.plot(capacities, MSE, label='MSE', marker='o')
plt.axvline(optimal_capacity, color='r', linestyle='--', label='Optimal Capacity')
plt.xlabel('Model Capacity')
plt.ylabel('Error')
plt.title('Bias-Variance Tradeoff')
plt.legend()
plt.show()
```

The reason why don't we use the real formula is because of the training time and training samples. The graph look terrible when we want to train and present in a short time
